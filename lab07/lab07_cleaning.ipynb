```python
df_clean = pd.DataFrame(data)

df_clean['wynagrodzenie'] = pd.to_numeric(df_clean['wynagrodzenie'].replace('brak', np.nan), errors='coerce')
df_clean = df_clean.drop_duplicates().reset_index(drop=True)

df_clean['wynagrodzenie'] = df_clean['wynagrodzenie'].fillna(df_clean['wynagrodzenie'].median())
df_clean['ocena_roczna'] = df_clean['ocena_roczna'].fillna(round(df_clean['ocena_roczna'].mean(), 2))

df_clean['imie'] = df_clean['imie'].str.strip().str.title()
df_clean['dzial'] = df_clean['dzial'].str.strip().str.title()
df_clean['dzial'] = df_clean['dzial'].str.replace('Hr', 'HR', regex=False)
df_clean['dzial'] = df_clean['dzial'].str.replace('It', 'IT', regex=False)

df_clean['data_zatrudnienia'] = pd.to_datetime(df_clean['data_zatrudnienia'], errors='coerce')

print("=== WERYFIKACJA ===")
print(f"Shape: {df_clean.shape}")
print(f"\nDtypes: \n{df_clean.dtypes}")
print(f"\nNaN per kolumna:\n{df_clean.isna().sum()}")
print(f"\nUnikalne działy: {sorted(df_clean['dzial'].unique())}")
print(f"\nPierwsze imiona: {df_clean['imie'].tolist()[:8]}")

print("\n=== PYTANIA BIZNESOWE ===")
print(df_clean.groupby('dzial')['wynagrodzenie'].mean().round(2))
print(df_clean.groupby('dzial')['ocena_roczna'].mean().round(2))
display(df_clean.nlargest(3, 'wynagrodzenie')[['imie', 'dzial', 'wynagrodzenie']])
print(df_clean['dzial'].value_counts())
```
