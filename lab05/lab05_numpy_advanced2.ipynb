## Zad 1a

```python
# Klucze: 'Anna', 'Jan', 'Ewa', 'Marek', 'Kasia'
# Wartości: 5500, 7200, 4800, 9100, 6300
data_dict = {
    'Anna': 5500, 
    'Jan': 7200, 
    'Ewa': 4800, 
    'Marek': 9100, 
    'Kasia': 6300
}
pensje = pd.Series(data_dict)
print(pensje)
```

## Zad 1b

```python
# 1. Wyświetl imię osoby zarabiającej najwięcej (użyj idxmax())
najbogatszy = pensje.idxmax()
print(najbogatszy)
# 2. Oblicz średnią pensję
średnia = pensje.mean()
print(średnia)
# 3. Wyświetl imiona osób zarabiających powyżej średniej (filtrowanie boolean)
powyzej = pensje > pensje.mean()
osoby_powyzej = pensje[powyzej]
print(osoby_powyzej.to_string())
# 4. Policz ile osób zarabia poniżej 6000 (użyj len() lub .sum())
ponizej = pensje < 6000
osoby_ponizej = pensje[ponizej]
print(len(osoby_ponizej))
```

## Zad 1c

```python
# 1. Oblicz pensje po podwyżce 10% dla wszystkich (pensje * 1.10)
podwyzka = pensje * 1,10
print("Pensje po podwyżce:")
print(podwyzka)
# 2. Oblicz pensje po premii 500 zł dla wszystkich (pensje + 500)
premia = pensje + 500
print("Pensje po premii:")
print(premia)
# 3. Oblicz nową pensję = (pensja * 1.10) + 500
nowa_pensja = (pensje * 1.10) + 500
print("Nowa pensja:")
print(nowa_pensja) 
# 4. Wyświetl ranking nowych pensji od najwyższej (użyj sort_values)
ranking = pensje.sort_values
print(ranking)
```

```python
tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
# Dodaj kolumnę z procentem napiwku
tips['tip_pct'] = (tips['tip'] / tips['total_bill'] * 100).round(1)
```

## Zad 4a

```python
# 1. Oblicz średni rachunek (total_bill)
sredni_rachunek = tips['total_bill'].mean()
print(f"1. Średni rachunek: {sredni_rachunek:.2f}")

# 2. Oblicz średni napiwek w procentach (tip_pct)
sredni_tip_pct = tips['tip_pct'].mean()
print(f"2. Średni napiwek (%): {sredni_tip_pct:.2f}%")

# 3. Wyświetl który dzień ma najwięcej rachunków
dni_counts = tips['day'].value_counts()
print("\n3. Liczba rachunków w poszczególne dni:")
print(dni_counts)
print(f"Najwięcej rachunków było w: {dni_counts.idxmax()}")

# 4. Policz ile rachunków jest od palaczy
palacze_counts = tips['smoker'].value_counts()
print(f"\n4. Liczba rachunków od palaczy (Yes): {palacze_counts['Yes']}")

# 5. Wyświetl describe() tylko dla kolumny 'tip' — jaka jest mediana napiwku?
print("\n5. Statystyki dla kolumny 'tip':")
print(tips['tip'].describe())
# Mediana to 50% w wynikach describe()
mediana_tip = tips['tip'].median()
print(f"Mediana napiwku wynosi: {mediana_tip}")

# 6. Jaki procent rachunków pochodzi z obiadu (Dinner)?
procent_czasu = tips['time'].value_counts(normalize=True) * 100
print("\n6. Procentowy podział czasu (Lunch vs Dinner):")
print(procent_czasu)
print(f"Procent rachunków z Dinner: {procent_czasu['Dinner']:.1f}%")

# 7. Ile unikalnych rozmiarów grup jest w danych?
unikalne_grupy = tips['size'].nunique()
print(f"\n7. Liczba unikalnych rozmiarów grup: {unikalne_grupy}")
print(f"Konkretne rozmiary grup: {tips['size'].unique()}")
```
